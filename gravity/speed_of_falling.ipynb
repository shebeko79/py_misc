```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from matplotlib import colors
```

```python
EARTH_RADIUS = 6371000
EARTH_MASS = 5.9722E24
G = 6.6743E-11
```

```python
%matplotlib notebook
```

```python
TOTAL_DIST=600000000*10
cur_v=0
cur_d=0
t=0
dt=1

while True:
    R=TOTAL_DIST-cur_d
    a=G*EARTH_MASS/(R*R)
    
    cur_v=cur_v+a*dt
    new_d=cur_d+cur_v*dt
    
    if new_d+EARTH_RADIUS > TOTAL_DIST:
        break
    
    cur_d=new_d
    t=t+dt
    
    if (t%(3600*24))==0:
        print(f'{t/3600/24}d a={a} V={cur_v} S={cur_d}')
        
print(f'final: {t/3600}h a={a} V={cur_v} S={cur_d}')

```

```python
LOM_MASS=5
E=LOM_MASS*cur_v*cur_v/2
print(f'Speed={cur_v}m/s m={LOM_MASS}kg E={E/1000000}MJ E={E/3600000}KWh')
```
